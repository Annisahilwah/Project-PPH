## Data cleaning

### Subtask:
Clean the data by handling missing values and outliers.


**Reasoning**:
Inspect the DataFrame for missing values, handle outliers in numerical features, and ensure correct data types.



```python
# Inspect missing values
print("Missing values per column:\n", df.isnull().sum())

# Handle missing values (no missing values found in the initial exploration)
# If there were missing values, imputation strategies would be applied here.
# For example:
# df['umur_ekonomis'].fillna(df['umur_ekonomis'].median(), inplace=True)

# Outlier analysis and handling for numerical features
numerical_cols = ['pendapatan', 'beban_operasional', 'penyusutan', 'tax_rate', 'nilai_perolehan', 'umur_ekonomis']
for col in numerical_cols:
    # Calculate IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in {col}:\n{outliers}")

    # Winsorizing (capping) outliers
    df[col] = np.clip(df[col], lower_bound, upper_bound)

# Ensure correct data types (check and convert if necessary)
# Example:
# if df['tahun'].dtype != 'int64':
#     df['tahun'] = df['tahun'].astype('int64')
#     print("Converted 'tahun' to int64.")
```

## Data analysis

### Subtask:
Perform statistical analysis on the data, including descriptive statistics, correlation analysis, and potentially hypothesis testing.


**Reasoning**:
Perform the statistical analysis as requested, including descriptive statistics, correlation analysis, and visualizations.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Descriptive Statistics
print("Descriptive Statistics:\n", df.describe())

# 2. Correlation Analysis
numerical_cols = ['pendapatan', 'beban_operasional', 'penyusutan', 'tax_rate', 'nilai_perolehan', 'umur_ekonomis', 'profitability_ratio', 'remaining_economic_life']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()


# 3. Relationship between categorical and numerical variables
plt.figure(figsize=(12, 6))
sns.boxplot(x='kategori', y='profitability_ratio', data=df, palette="Set3")
plt.title('Profitability Ratio by Asset Category')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='metode', y='pendapatan', data=df, palette="Set3")
plt.title('Revenue by Depreciation Method')
plt.xticks(rotation=45)
plt.show()

# 4. Hypothesis Testing (Example: Relationship between remaining_economic_life and profitability_ratio)
from scipy.stats import pearsonr

# Null Hypothesis: There is no linear correlation between remaining_economic_life and profitability_ratio.
# Alternative Hypothesis: There is a linear correlation between remaining_economic_life and profitability_ratio.
correlation_coefficient, p_value = pearsonr(df['remaining_economic_life'], df['profitability_ratio'])

print(f"\nPearson Correlation Coefficient: {correlation_coefficient:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant correlation.")
else:
    print("Fail to reject the null hypothesis. No statistically significant correlation found.")
```
